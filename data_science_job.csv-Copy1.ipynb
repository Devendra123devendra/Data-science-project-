# find the missing value column bise

```python
df.isnull().mean()*100
```

# select column name (below 5%)

```python

threshold = 0.05  # 5%

# Calculate missing value percentage per column
missing_percent = df.isnull().mean()

# Select column names with missing % > 0 and < 5%
selected_columns = [
    col for col in df.columns
    if 0 < missing_percent[col] < threshold
]

print(selected_columns)
```

```python
df[selected_columns].sample(5)
```

# calculate the drop data

```python
len(df[selected_columns].dropna())/len(df)
```

# create a new data

```python
new_df = df[selected_columns].dropna()
df.shape,new_df.shape
```

# plot histogram (before applying cca: Numerical data

```python
new_df.hist(bins=50,density=True,figsize=(12,12))
plt.show()
```

# plot histogram training hours

```python

fig = plt.figure()
ax = fig.add_subplot(111)

# original data
df['training_hours'].hist(
    bins=50,
    ax=ax,
    density=True,
    color='red'
)

# data after CCA (overlay)
new_df['training_hours'].hist(
    bins=50,
    ax=ax,
    density=True,
    color='green',
    alpha=0.8
)

plt.xlabel('Training Hours')
plt.ylabel('Density')
plt.title('Training Hours Distribution (Before vs After CCA)')
plt.show()
```

# plot problity density function (pdf): Treaining hours

```python
fig = plt.figure()
ax = fig.add_subplot(111)

# original data
df['training_hours'].plot.density(color='red', ax=ax)

# data after CCA
new_df['training_hours'].plot.density(color='green', ax=ax)

plt.xlabel('Training Hours')
plt.ylabel('Density')
plt.title('Training Hours Density Plot (Before vs After CCA)')
plt.legend(['Original', 'After CCA'])
plt.show()
```
